```python
import pandas as pd
import json

# 1. 读取 Excel 文件
file_path = "low_complete.xlsx"
df = pd.read_excel(file_path)

# 2. 将每条记录转为字典
records = df.to_dict(orient="records")  # 等价于逐行转换

# 3. 保存为 jsonl（一行一个 JSON 对象）
output_path = "./dataset/low_complete_v1.jsonl"
with open(output_path, "w", encoding="utf-8") as f:
    for row in records:
        f.write(json.dumps(row, ensure_ascii=False) + "\n")

print("转换完成，JSONL 文件已生成:", output_path)
```
